```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/proje_train.csv")
df
```

```python
# prompt: calculate body mass index

df["bmi"] = df["Weight_kg"] / (df["Height_cm"] / 100)**2
df
```

```python
# prompt: classificate according to bmi

import pandas as pd
df["BMI_Category"] = pd.cut(df["bmi"], bins=[0, 18.5, 25, 30, 35, 40, 45, 50, 55],
                           labels=["Underweight", "Normal weight", "Overweight", "Obesity Class I",
                                   "Obesity Class II", "Obesity Class III", "Obesity Class IV", "Obesity Class V"])

df
```

```python
# prompt: unique values of participant id column in df

unique_values = df.nunique()
print(unique_values)
```

```python
df.isna().sum()
```

```python
# prompt: every value of columns

for column in df:
  unique_values = df[column].unique()
  print(f"Unique values in column {column}: {unique_values}")
```

```python
# prompt: Change Omni to Omnivore in Diet_Type column

df['Diet_Type'] = df['Diet_Type'].replace('Omni', 'Omnivore')
df['Diet_Type'] = df['Diet_Type'].replace('veg*n', 'Vegetarian')
df['Diet_Type'] = df['Diet_Type'].replace('Vegan', 'Vegetarian')
```

```python
diet_types = df['Gender'].unique()

for diet_type in diet_types:
  print(diet_type)
```

```python
# prompt: change decimal to 2 in  Hours_Sleep_Per_Night column

df['Hours_Sleep_Per_Night'] = df['Hours_Sleep_Per_Night'].apply(lambda x: round(x, 2))
```

```python
df['Height_cm'] = df['Height_cm'].apply(lambda x: round(x, 2))
```

```python
df['Weight_kg'] = df['Weight_kg'].apply(lambda x: round(x, 2))
```

```python
# prompt: change rows that has -1 value on Stress_Level column to 0

df.loc[df['Stress_Level'] == -1, 'Stress_Level'] = 0
```

```python
# prompt: average of Stress_Level of diet types

df_grouped = df.groupby('Diet_Type')['Stress_Level'].mean()
print(df_grouped)
```

```python
# prompt: average of Smoking status of Stress_Level

df_grouped = df.groupby('Smoking_Status')['Stress_Level'].mean()
print(df_grouped)
```

```python
# prompt: drop "PID_" part from Partipicant_ID column

df['Participant_ID'] = df['Participant_ID'].str.replace('PID_', '')
df['Participant_ID'] = df['Participant_ID'].astype("int")
```

```python
# prompt: draw boxplot for every value in Smoking_Status with Stress_Level

import seaborn as sns

sns.boxplot(x="Smoking_Status", y="Stress_Level", data=df)
```

```python
# prompt: average of Daily_Fruit_Consumption of Overall_Health_Rating

average_daily_fruit_consumption_by_overall_health_rating = df.groupby('Daily_Fruit_Consumption')['Overall_Health_Rating'].mean()
print(average_daily_fruit_consumption_by_overall_health_rating)
```

```python
# prompt: make one hot encoding to gender column

import pandas as pd
df = pd.get_dummies(df, columns=['Gender'], dtype=int)
```

```python
# prompt: make one hot encoding to diet_type column

import pandas as pd
df = pd.get_dummies(df, columns=['Diet_Type'], dtype=int)
```

```python
# prompt: give correlation map of the data

import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr()

fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corr, annot=True, fmt=".2f", ax=ax, cmap="coolwarm", square=True)
plt.show()
```
